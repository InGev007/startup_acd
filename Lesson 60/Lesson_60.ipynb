```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

## Функції завантаженння та збереження картинки

```python
# Функція для завантаження зображеня і відображення на екрані
def load(path):
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(img)
    return img
     
# Функція запису зображення на диск
def writing(img, path="img.jpg"):
    cv2.imwrite(path, img)
     
```

##Завантаження картинки

```python
img = load('/content/aidared.jpg')
```

##Використання фільтру Собеля

```python
sobel = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=1)
plt.axis('off')
plt.imshow(sobel)
```

```python
# Застосування фільтру Canny для виділення горизонтальних ліній для кольорового зображення
canny = cv2.Canny(img,0,200)
plt.axis('off')
plt.imshow(canny)
```

```python
roberts_kernel_v = np.array([[1, 0], [0, -1]], dtype=np.float32)  # вертикальні границі
roberts_kernel_g = np.array([[0, 1], [-1, 0]], dtype=np.float32)  # горизонтальні границі
# Застосування фільтру Робертса 
roberts_v = cv2.filter2D(img, -1, roberts_kernel_v)
roberts_g = cv2.filter2D(img, -1, roberts_kernel_g)

roberts = roberts_v + roberts_g

plt.axis('off')
plt.imshow(roberts)
```
