Побудувати дискретний статистичний розподіл для цієї вибірки, а також полігон частот і F*(x). Знайти моду та медіану.

## Вирішення:

####  Cпочатку потрібно побудувати дискретний статистичний розподіл для даної вибірки:

```python
x = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]

frequency = {}
for i in x:
    if i in frequency:
        frequency[i] += 1
    else:
        frequency[i] = 1

n = len(x)

for key in frequency:
    frequency[key] = frequency[key]/n

print("Дискретний статистичний розподіл:")
for key, value in frequency.items():
    print(key, ":", value)
```

#### Побудуємо полігон частот:

```python
import matplotlib.pyplot as plt

plt.bar(frequency.keys(), frequency.values(), width=0.5)
plt.xlabel("X")
plt.ylabel("Частота")
plt.title("Полігон частот")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sorted_sample = np.sort(x)
ecdf = np.arange(1, len(sorted_sample) + 1) / len(sorted_sample)

plt.step(sorted_sample, ecdf)
plt.xlabel('X')
plt.ylabel('F*(x)')
plt.title('Empirical CDF')
plt.show()
```

#### Визначимо моду
Мода - це значення вибірки, яке зустрічається найчастіше. В нашому випадку, мода дорівнює 9, оскільки це значення зустрічається найбільшу кількість разів.

```python
from collections import Counter

# використовуємо Counter для підрахунку кількості входжень кожного елемента
counted = Counter(x)

# знаходимо найбільше значення
mode = max(counted, key=counted.get)

print("Мода:", mode)
```

#### Визначимо Медіану
Медіана - це значення, яке знаходиться в середині впорядкованої вибірки. Для знаходження медіани, спочатку потрібно впорядкувати вибірку:

```python
x.sort()
n = len(x)
if n % 2 == 0:
    median1 = x[n//2]
    median2 = x[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = x[n//2]

print("Медіана:", median)
```
