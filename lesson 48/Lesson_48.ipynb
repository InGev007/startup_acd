##Імпортуємо бібліотеки та підключаємо данні

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Data_to_Transform.csv')
df.head()
```

##Проводимо трансформації для Moderate Positive Skew

```python
sns.histplot(df['Moderate Positive Skew'])
fig = plt.figure()
```

```python
df_log = np.log(df['Moderate Positive Skew'])
sns.histplot(df_log)
fig = plt.figure()
```

```python
df_cbrt = np.cbrt(df['Moderate Positive Skew'])
sns.histplot(df_cbrt)
fig = plt.figure()
```

```python
data = df['Moderate Positive Skew'].values.reshape(-1, 1)
pt = PowerTransformer(method='box-cox')
df_boxcox = pt.fit_transform(data)
sns.histplot(df_boxcox)
fig = plt.figure()
```

```python
df1 = np.int64(df['Moderate Positive Skew'])
fig = sm.qqplot(df1, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_log, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_cbrt, line='45')
plt.show()
```

```python
fig = sm.qqplot(df_boxcox, line='45')
plt.show()
```
