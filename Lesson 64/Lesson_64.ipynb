```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im_dst = cv2.imread('chess_1.png')
pts_dst = np.array([[465, 39],[460, 261],[200, 245],[190, 50]])


plt.imshow(im_dst)
plt.axis('off')
plt.show()
```

```python
im_src = cv2.imread("chess_2.png")
pts_src = np.array([[166, 33], [530, 100], [409, 364],[189, 317]])


plt.imshow(im_src)
plt.axis('off')
plt.show()
```

```python
h, status = cv2.findHomography(pts_src, pts_dst)

im_out = cv2.warpPerspective(im_src,
                             h,
                             (im_dst.shape[1], im_dst.shape[0]))



plt.imshow(im_out)
plt.axis('off')
plt.show();
```
