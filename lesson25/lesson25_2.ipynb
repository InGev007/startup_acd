## Завдання 2:

Розглянемо дві різні умовні акції X і Y. Для кожної з них за допомогою статистичних досліджень були розраховані відносні частоти очікуваних прибутків залежно від п'яти станів економічного середовища: значне піднесення(ЗП), незначне піднесення(НЗП), стагнація(С), незначний спад(НЗС), значний спад(ЗС). Ці дані занесені відповідно в останній рядок і останній стовпчик таблиці. Залежно від економічного середовища прогнозовані норми очікуваних прибутків у відсотках задані відповідно у першому рядку та першому стовпчику таблиці. У клітинах на перетині X i Y проставлені ймовірності появи норм відсотків значень (х, у).

| **Y / X** |10      | 5      | 2      | 1      | -5     | P(Y)  |
| :---      | :----: | :----: | :----: | :----: | :----: | ---:  |
| **20**    | 0,01   | 0,03   | 0,02   | 0,03   |0,01    |**0,1**|
| **10**    | 0,03   | 0,09   | 0,06   | 0,09   |0,03    |**0,3**|
| **2**     | 0,02   | 0,06   | 0,04   | 0,06   |0,02    |**0,2**|
| **-2**    | 0,03   | 0,09   | 0,06   | 0,09   |0,03    |**0,3**|
| **-8**    | 0,01   | 0,03   | 0,02   | 0,03   |0,01    |**0,1**|
| **P(X)**  | **0,1**|**0,3** |**0,2** |**0,3** |**0,1** |**1**  |

Необхідно:

1) обчислити очікувану норму прибутку по кожній з акцій;
2) оцінити ступінь ризику обох акцій за допомогою середньоквадратичного відхилення;
3) проаналізувати наявність зв'язку між акціями X та Y за допомогою коефіціентів коваріації та кореляції.

## Вирішення:

#### Обчислюємо математичне сподівання для X та Y 
$M[X] = \sum_{i=1}^∞X_i P_i $

```python
Mx=10*0.1 + 5*0.3 + 2*0.2 + 1*0.3 - 5*0.1
My=20*0.1 + 10*0.3 + 2*0.2 - 2*0.3 - 8*0.1
print(f"Математичне сподівання для X: {round(Mx,2)}")
print(f"Математичне сподівання для Y: {round(My,2)}")
```

#### Знаходимо дисперсію для X та Y:
$D(X) =  M(X^2) - M(X)^2$

```python
Dx=(10**2 * 0.1 + 5**2 * 0.3 + 2**2 * 0.2 + 1**2 * 0.3 + (- 5)**2 * 0.1)  - Mx**2
Dy=(20**2 * 0.1 + 10**2 * 0.3 + 2**2 * 0.2 + (- 2)**2 * 0.3 + (- 8)**2 * 0.1)  - My**2
print(f"Дисперсія для X: {round(Dx,2)}")
print(f"Дисперсія для Y: {round(Dy,2)}")
```

#### Знаходимо середньоквадратичне відхилення для X та Y
$\delta (X) = \sqrt{D(X)}$

```python
from math import sqrt
SDx=sqrt(Dx)
SDy=sqrt(Dy)
print(f"Cередньоквадратичне відхилення для X: {round(SDx,2)}")
print(f"Cередньоквадратичне відхилення для Y: {round(SDy,2)}")
```

#### Знаходимо коваріацію
$ cov(X,Y) = M(XY) - M(X) \bullet M(Y) $

```python
Mxy=10*(20*0.01 + 10*0.03 + 2*0.02 - 2*0.03 - 8*0.01) + 5*(20*0.03 + 10*0.09 + 2*0.06 - 2*0.09 - 8*0.03) + 2*(20*0.02 + 10*0.06 + 2*0.04 - 2*0.06 - 8*0.02) + 1*(20*0.03 + 10*0.09 + 2*0.06 - 2*0.09 - 8*0.03) - 5*(20*0.01 + 10*0.03 + 2*0.02 - 2*0.03 - 8*0.01)
```

```python
covXY=Mxy-Mx*My
print(f"Коваріація: {round(covXY,2)}")
```

#### Знаходимо кореляцію
$ r_{xy} = \frac {cov(X,Y)}{\delta (X)\delta (Y)}$ 

```python
Rxy=covXY/(SDx*SDy)
print(f"Кореляція: {round(Rxy,2)}")
```

```python
print(f"Очікувана норму прибутку по X = {round(Mx,2)}, а для акції Y = {round(My,2)}. Акція Y більш прибуткова.")
print(f"Cередньоквадратичне відхілення у акцій Х менше, то вони ризиковані")
print(f"Звязку між акціями X та Y немає")
```
