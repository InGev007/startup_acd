```python
import numpy as np
import cv2 as cv2
import matplotlib.pyplot as plt
```

```python
# Функція для завантаження зображеня і відображення на екрані
def load(path):
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(img)
    return img
```

```python
# Функція запису зображення на диск
def writing(img, path="img.jpg"):
    cv2.imwrite(path, img)
```

```python
def change_contrast(image, contrast_factor):
    # Конвертирование изображения в формат cv2
    image_cv2 = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    # Применение контраста
    adjusted_image = cv2.convertScaleAbs(image_cv2, alpha=contrast_factor)
    # Конвертирование обратно в формат RGB
    adjusted_image = cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(adjusted_image)
    return adjusted_image
```

```python
def change_brightness(image, value):
    # Конвертирование изображения в формат cv2
    image_cv2 = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    # Применение линейного преобразования к яркости
    alpha = float(value) / 255.0
    adjusted_image = cv2.convertScaleAbs(image_cv2, alpha=alpha, beta=0)
    # Конвертирование обратно в формат RGB
    adjusted_image = cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(adjusted_image)
    return adjusted_image
```

```python
def change_saturation(image, saturation_factor):
    # Конвертирование изображения в формат cv2
    image_cv2 = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    # Преобразование изображения в HSV цветовое пространство
    hsv_image = cv2.cvtColor(image_cv2, cv2.COLOR_BGR2HSV)
    # Изменение насыщенности изображения
    hsv_image[..., 1] = np.clip(hsv_image[..., 1] * saturation_factor, 0, 255)
    # Конвертирование обратно в формат RGB
    adjusted_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB)
    plt.axis('off')
    plt.imshow(adjusted_image)
    return adjusted_image
```

```python
def change_size(image, width=None, height=None, scale=None):
    # Конвертирование изображения в формат cv2
    image_cv2 = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    # Получение размера изображения
    img_height, img_width = image_cv2.shape[:2]
    # Вычисление нового размера изображения
    if scale:
        new_height = int(img_height * scale)
        new_width = int(img_width * scale)
    else:
        new_height = height
        new_width = width
    # Изменение размера изображения
    adjusted_image = cv2.resize(image_cv2, (new_width, new_height))
    # Конвертирование обратно в формат RGB
    adjusted_image = cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(adjusted_image)
    return adjusted_image
```

```python
def crop_image(image, x, y, width, height):
    # Конвертирование изображения в формат cv2
    image_cv2 = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    # Выделение необходимой области изображения
    cropped_image = image_cv2[y:y+height, x:x+width]
    # Конвертирование обратно в формат RGB
    cropped_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(cropped_image)
    return cropped_image
```

```python
def rotate_image(image, angle):
    # Конвертирование изображения в формат cv2
    image_cv2 = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    # Получение размера изображения
    height, width = image_cv2.shape[:2]
    # Определение матрицы поворота
    matrix = cv2.getRotationMatrix2D((width/2, height/2), angle, 1)
    # Применение матрицы поворота к изображению
    rotated_image = cv2.warpAffine(image_cv2, matrix, (width, height))
    # Конвертирование обратно в формат RGB
    rotated_image = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(rotated_image)
    return rotated_image
```
