## Завдання 1:

Мікропроцесор має 10000 транзисторів, які працюють незалежно один від одного. Імовірність того, що транзистор вийде з ладу під час роботи приладу, є величиною малоймовірною і становить 0,0007. Визначити математичне сподівання М (Х) і середнє квадратичне відхилення S (Х) випадкової величини Х — числа транзисторів, що вийдуть із ладу під час роботи мікропроцесора.

## Вирішення:

Для вирішення цього завдання можна скористатися формулами для математичного сподівання та середнього квадратичного відхилення випадкової величини.

Математичне сподівання M(X) можна знайти за формулою:
$M[X] = n * P$

де n - кількість спостережень, тобто кількість транзисторів в мікропроцесорі, p - імовірність того, що транзистор вийде з ладу.

Середнє квадратичне відхилення S(X) можна знайти за формулою:
$S(X) = sqrt(n * p * (1 - p))$

де sqrt() - функція обчислення квадратного кореня.

Отже, для даної задачі математичне сподівання та середнє квадратичне відхилення випадкової величини можна обчислити таким чином:


```python
from math import sqrt

n = 10000
p = 0.0007

M = n * p
S = sqrt(n * p * (1 - p))

print("Математичне сподівання: ", M)
print("Середнє квадратичне відхилення: ", S)
```
