## Завдання 2:

У рибальському містечку 99,99% чоловіків хоча б раз в житті були на рибалці. Проводять соціологічні дослідження серед 10000 навмання вибраних чоловіків. Визначити дисперсію D (X) і середнє квадратичне відхилення S (Х) випадкової величини Х — числа чоловіків, які жодного разу не були на рибалці.

## Вирішення:

З огляду на те, що величина Х визначає кількість чоловіків, які не були на рибалці, то можна припустити, що вона є біноміально розподіленою випадковою величиною з параметрами $n = 10000$ (загальна кількість чоловіків у дослідженні) і $p = 0.0001$ (імовірність того, що чоловік ніколи не був на рибалці).

Отже, для визначення дисперсії та середнього квадратичного відхилення випадкової величини X можна використати формули для біноміального розподілу:

$D(X) = np(1-p)$

$S(X) = sqrt(np(1-p))$

де n - кількість спостережень, p - імовірність успіху.

У нашому випадку:

$n = 10000$
$p = 0.0001$

Тоді код на Python буде наступним:

```python
from math import sqrt

n = 10000
p = 0.0001

D = n * p * (1 - p)
S = sqrt(D)

print("Дисперсія: ", D)
print("Середнє квадратичне відхилення: ", S)
```
