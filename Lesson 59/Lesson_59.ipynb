```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import warnings
warnings.filterwarnings("ignore")
```

## Функції завантаженння та збереження картинки

```python
# Функція для завантаження зображеня і відображення на екрані
def load(path):
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(img)
    return img
     
# Функція запису зображення на диск
def writing(img, path="img.jpg"):
    cv2.imwrite(path, img)
     
```

##Завантаження картинки

```python
img = load('/content/aidared.jpg')
```

##Використання фільтру Собеля

```python
sobel = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=1)
plt.axis('off')
plt.imshow(sobel)
```

```python
sobel = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
plt.axis('off')
plt.imshow(sobel)
```

```python
sobel = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=9)
plt.axis('off')
plt.imshow(sobel)
```

```python
sobel = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=9)
plt.axis('off')
plt.imshow(sobel)
```

##Використання Медіанного фільтру

```python
median  = cv2.medianBlur(img,3)
plt.axis('off')
plt.imshow(median)
```

```python
median  = cv2.medianBlur(img,5)
plt.axis('off')
plt.imshow(median)
```

```python
median  = cv2.medianBlur(img,9)
plt.axis('off')
plt.imshow(median)
```
