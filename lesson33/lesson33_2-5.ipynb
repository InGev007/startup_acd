2. Візуалізуйте залежності: споживання палива від вартості – лінійною діаграмою; споживання палива від швидкості – точковим графіком; вантажопідйомності від швидкості – точковим графіком.

```python
import pandas as pd


# Імпортуємо датафрейм
df = pd.read_excel('автомобілі.xls')

df.head(3)
```

```python
import matplotlib.pyplot as plt


# Побудова графіку
plt.plot(df['Вартість, тис. дол.'], df['Споживання палива, л/100 км'])

# Додаткові налаштування графіку
plt.title('Залежність споживання палива від вартості автомобіля')
plt.xlabel('Вартість, $')
plt.ylabel('Споживання палива, л/100 км')
#plt.xticks(df['Вартість, тис. дол.'], df['Марка автомобіля'], rotation=90)
plt.grid(True)

# Відображення графіку
plt.show()
```

```python
import seaborn as sns

sns.lineplot(data=df, x='Вартість, тис. дол.', y='Споживання палива, л/100 км')
```

```python
# Побудова графіку
plt.scatter(df['Споживання палива, л/100 км'], df['Максимальна швидкість, км/год'], c = 'deeppink')

# Додаткові налаштування графіку
plt.title('Залежність споживання палива від швидкості автомобіля')
plt.xlabel('Споживання палива, л/100км')
plt.ylabel('Максимальна швидкість, км/год')
plt.grid(True)

# Відображення графіку
plt.show()
```

```python
# Побудова графіку
plt.scatter(df['Вантажопідйомність, т'], df['Споживання палива, л/100 км'], c = 'deeppink')

# Додаткові налаштування графіку
plt.title('Залежність вантажопідйомності від споживання палива автомобіля')
plt.xlabel('Вантажопідйомність, т')
plt.ylabel('Споживання палива, л/100 км')
plt.grid(True)

# Відображення графіку
plt.show()
```

3. Побудуйте гістограму для розділення автомобілів на 5 груп за швидкістю.

```python
# Побудова гістограми
plt.hist(df['Максимальна швидкість, км/год'], bins=5)
plt.xlabel('Швидкість, км/год')
plt.ylabel('Кількість автомобілів')
plt.title('Розподіл автомобілів за швидкістю')
plt.show()
```

```python
# Побудувати коробковий графік 
plt.boxplot(df['Максимальна швидкість, км/год'])
plt.ylabel('Max Speed')
plt.title('Speed Distribution by Brand')
plt.show()
```
